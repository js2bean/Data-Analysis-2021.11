# 데이터 시각화 - Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 라인플롯

plt.plot([0,1,2,3],[0,1,4,9])       # x값 리스트, y값 리스트
plt.title('Line plot')              # 제목
plt.grid()                          # 격자 라인
plt.show()                          # 합쳐서 화면에 보이게 함
```
