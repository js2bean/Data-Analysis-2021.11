```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot([0,1,2,3],[0,1,4,9])       
plt.title('라인 플롯')              
plt.grid()                          
plt.show()  
```

- Line style

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')        #color, marker, linestyle
plt.title('Red-Circle-Line')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--')        #color, marker, linestyle
plt.plot([8,4,2,1], 'g^:')
plt.title('Magenta-suare-dashed, Green-Triangle-dotted')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--', linewidth=3, ms=10)
plt.plot([8,4,2,1], 'g^:', lw=5, ms=20)
plt.title('Magenta-suare-dashed, Green-Triangle-dotted')
plt.grid()
plt.show()
```

- 그림 범위 지정

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')
plt.title('x축, y축 범위지정')
plt.xlim(-1,4)
plt.ylim(-1,10)
plt.grid()
plt.show()
```

- Tick 설정

```python
X = np.linspace(-np.pi, np.pi, 500)
Y = np.cos(X)
plt.title('Tick 설정')
plt.plot(X, Y)
plt.xticks([-np.pi, -np.pi/2, 0 , np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.show
```

- 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한 플롯')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0 , np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.show
```

- X축, Y축 라벨

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('X축, Y축 라벨을 표시한 플롯')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0 , np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show
```

- 그림의 크기

```python
plt.figure(figsize=(10,6))
plt.title('그림 크기 지정')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0 , np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.show
```

- 그림 저장

```python
plt.figure(figsize=(10,6))
plt.title('저장')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0 , np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.savefig('삼각함수.png')
```
