- 결측치 처리

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isna().sum()
```

```python
titanic.isna().sum().sum()
```

```python
titanic.tail()
```

```python
# age 결측치는 평균으로 대체

titanic['age'].fillna (titanic.age.mean(), inplace=True)
titanic.tail(3)
```

```python
# embarked 결측치는 행 삭제
titanic.dropna(subset=['embarked'], inplace=True)
titanic.dropna(subset=['embark_town'], inplace=True)
titanic.tail()
```

```python
# deck 컬럼은 컬럼삭제

titanic.drop(columns=['deck'], inplace=True)
titanic.tail()
```

- apply()

```python
df
```

```python
# 각 열의 최대값과 최소값의 차이

df.apply(lambda x: x.max() - x.min())       #axis=0가 생략
```

```python
# 각 행의 최대값과 최소값의 차이
df.apply(lambda x: x.max() - x.min(), axis=1)  
```

```python
# 타이타닉호 승객중 20살이 넘으면 'adult' 그렇지 않으면 'child'lambda

titanic['adult/child'] = \
    titanic.apply(lambda x: 'adult' if x.age>20 else 'child', axis=1)
titanic.tail()
```
